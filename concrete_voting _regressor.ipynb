```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
df = pd.read_csv("C:/Users/User/Desktop/cdac advance modules/pml lecture/class  practice/program_dataset/concrete_data.csv")
```

```python
df
```

```python

from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.ensemble import VotingRegressor
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, 
                                                    random_state=42)
```

```python
dtr = DecisionTreeRegressor(max_depth=4)
lr = LinearRegression()
ridge = Ridge()
```

```python
dtr.fit(X_train,y_train)
y_predDT = dtr.predict(X_test)
r1=mean_squared_error(y_test, y_predDT) ** 0.5
print(mean_squared_error(y_test, y_predDT) ** 0.5)

lr.fit(X_train,y_train)
y_predLR = lr.predict(X_test)
r2=mean_squared_error(y_test, y_predLR) ** 0.5
print(mean_squared_error(y_test, y_predLR) ** 0.5)

ridge.fit(X_train,y_train)
y_predR = ridge.predict(X_test)
r3=mean_squared_error(y_test, y_predR) ** 0.5
print(mean_squared_error(y_test, y_predR) ** 0.5)

print("w1= ",(r1/(r1+r2+r2)))
print("w2= ",(r2/(r1+r2+r2)))
print("w3= ",(r3/(r1+r2+r2)))
```

```python

# Average
Voting = VotingRegressor(estimators=[('DT',dtr),
                                     ('LR',lr),('RD',ridge)])


Voting.fit(X_train,y_train)
y_pred = Voting.predict(X_test)


print(mean_squared_error(y_test, y_pred) ** 0.5)
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python

#OR Weighted Average
Voting = VotingRegressor(estimators=[('DT',dtr),
                                     ('LR',lr),('RD',ridge)],
                                     weights=np.array([0.33,0.33,
                                                       0.33]))

Voting.fit(X_train,y_train)
y_pred = Voting.predict(X_test)


print(mean_squared_error(y_test, y_pred) ** 0.5)
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
