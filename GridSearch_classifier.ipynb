```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv(r"C:/Users/User/Desktop/cdac advance modules/pml lecture/class  practice/program_dataset/BreastCancer.csv")
```

```python
dum_df = pd.get_dummies(df1,drop_first=True)
dum_df
```

```python
X1 = dum_df.iloc[:,:-1]
y1 = dum_df.iloc[:,-1]
```

```python
knnc = KNeighborsClassifier()
svc = SVC()
```

```python
for clf in(knnc,svc):
    if clf==knnc:
        parameters = {'n_neighbors': np.arange(1,16)}
        kfold = KFold(n_splits=5 , random_state=42,shuffle=True)
        cv = GridSearchCV(knnc, param_grid=parameters,
                  cv=kfold,scoring='roc_auc',verbose=0)
        cv.fit( X1 , y1 )
        print(clf.__class__.__name__,cv.best_params_)
        print(clf.__class__.__name__,cv.best_score_)

    elif clf==svc:
        C_range = np.linspace(0.001,2,10)
        param_grid = dict( C=C_range)
        from sklearn.model_selection import StratifiedKFold
        kfold = StratifiedKFold(n_splits=5, random_state=2020,shuffle=True)
        cv= GridSearchCV(SVC(probability=True,kernel='linear'), 
                       param_grid=param_grid, cv=kfold, 
                       scoring="roc_auc")
        cv.fit( X1 , y1 )
        print(clf.__class__.__name__,cv.best_params_)
        print(clf.__class__.__name__,cv.best_score_)
        
```
