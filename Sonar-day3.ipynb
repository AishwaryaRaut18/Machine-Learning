```python
import pandas as pd

df = pd.read_csv("C:/Users/User/Desktop/cdac advance modules/pml lecture/class  practice/program_dataset/Sonar.csv")
```

```python
from sklearn.model_selection import StratifiedKFold
#from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.naive_bayes import MultinomialNB

X = df.iloc[:,:-1]
y = df.iloc[:,-1]

#kfold = KFold(n_splits=5, random_state=42)
kfold = StratifiedKFold(n_splits=5, random_state=2021,
                        shuffle=True)
multinomial = MultinomialNB()
# cross_val_score()
## 1. Fits the model for every fold combination
## 2. Evaluates the model fitted for every fold combination
results = cross_val_score(multinomial, X, y, cv=kfold, 
                          scoring='roc_auc')
print(results)
print("ROC AUC: %.4f (%.4f)" % (results.mean(), results.std()))
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, accuracy_score
from sklearn.neighbors import KNeighborsClassifier


X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.3, 
                                                    random_state=2021,
                                                    stratify=y)


knn = KNeighborsClassifier(n_neighbors=5)

knn.fit( X_train , y_train )
y_pred = knn.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))

```

```python
#########################################################
from sklearn.metrics import roc_curve, roc_auc_score

# Compute predicted probabilities: y_pred_prob
y_probs = knn.predict_proba(X_test)
y_pred_prob = y_probs[:,1]
# Generate ROC curve values: fpr, tpr, thresholds
#fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

roc_auc_score(y_test, y_pred_prob)
```
