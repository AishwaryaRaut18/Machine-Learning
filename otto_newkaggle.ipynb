```python
X=df.iloc[:,1:-1]
y=df.iloc[:,-1]
test=test1.iloc[:,1:]
```

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X,y)
y_probs = logreg.predict_proba(test)
```

```python
ID=test1['id']
```

```python
submit3=pd.DataFrame(y_probs,index=ID,columns=sampsub.columns[1:])
submit3.to_csv("./submitnewkaggleLR1.csv")
```

# Lda

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
da = LinearDiscriminantAnalysis()
da.fit(X,y)
y_pred_prob = da.predict_proba(test)
```

```python
submit2=pd.DataFrame(y_pred_prob,index=ID,columns=sampsub.columns[1:])
submit2.to_csv("./submitnewkaggleLDA.csv")
```

# QDA

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
da = QuadraticDiscriminantAnalysis()
da.fit(X,y)
y_pred_prob = da.predict_proba(test)
```

```python
submit1=pd.DataFrame(y_pred_prob,index=ID,columns=sampsub.columns[1:])
submit1.to_csv("./submitnewkaggleQDA.csv")
```

# DT

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=2021)
clf = clf.fit(X,y)
y_probs = clf.predict_proba(test)
```

```python
submit=pd.DataFrame(y_pred_prob,index=ID,columns=sampsub.columns[1:])
submit.to_csv("./submitnewkaggleDT.csv")
```
