```python
import numpy as np
import pandas as pd

df = pd.read_csv("C:/Users/User/Desktop/cdac advance modules/pml lecture/class  practice/program_dataset/Concrete_Data.csv")
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Import the necessary modules
from sklearn.model_selection import train_test_split 

# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, 
                                                    random_state=2018)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

from sklearn.linear_model import Ridge
ridge = Ridge()

from sklearn.linear_model import Lasso
lasso=Lasso()

from sklearn.svm import SVR
svr = SVR()

from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
```

```python
from sklearn.ensemble import StackingRegressor
models_considered = [('Linear Regression', lr),
                     ('Ridge', ridge),('Lasso',lasso),('SVR',svr),
                     ('Decision Tree',dtr)]
```

```python
from xgboost import XGBRegressor
clf = XGBRegressor()

stack = StackingRegressor(estimators = models_considered,
                           final_estimator=clf)

stack.fit(X_train,y_train)
```

```python
y_pred = stack.predict(X_test)
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```

```python
######## Include the original data along with predicted variables #####

stack = StackingRegressor(estimators = models_considered,
                           final_estimator=clf,
                           passthrough=True)

stack.fit(X_train,y_train)
```

```python
######################## Test set Operations ##########################
y_pred = stack.predict(X_test)
print(r2_score(y_test, y_pred))
```
