```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("C:/Users/User/Desktop/cdac advance modules/pml lecture/class  practice/program_dataset/Sonar.csv")
```

```python
dum_df = pd.get_dummies(df,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, accuracy_score,roc_auc_score
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from xgboost import XGBClassifier
```

```python
X = dum_df.iloc[:,:-1]
y = dum_df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.3, 
                                                    random_state=2021,
                                                    stratify=y)
```

```python
svc=SVC(probability=True)
BaggingClassifer = BaggingClassifier(random_state=2021,oob_score=True,
                             max_features=X_train.shape[1],
                             n_estimators=15,
                             max_samples=X_train.shape[0])

XGBClassifier=XGBClassifier()
```

```python
for clf in(svc,BaggingClassifer,XGBClassifier):
    if clf==svc:
        svc.fit(X_train, y_train) 
        y_probs = svc.predict_proba(X_test) 
        y_pred = svc.predict(X_test) 
        print(clf.__class__.__name__,(accuracy_score(y_test, y_pred)))
        print("ROC_AUC of ",clf.__class__.__name__,(roc_auc_score(y_test, y_pred)))
    elif clf==BaggingClassifer:
        BaggingClassifer.fit(X_train,y_train)
        y_pred = BaggingClassifer.predict(X_test)
        print(clf.__class__.__name__,(accuracy_score(y_test, y_pred)))
        print("ROC_AUC of ",clf.__class__.__name__,(roc_auc_score(y_test, y_pred)))
    elif clf==XGBClassifier:
        XGBClassifier.fit(X_train,y_train)
        y_pred = XGBClassifier.predict(X_test)
        print(clf.__class__.__name__,(accuracy_score(y_test, y_pred)))
        print("ROC_AUC of ",clf.__class__.__name__,(roc_auc_score(y_test, y_pred)))
    else:
        print("classifier not found")
```
